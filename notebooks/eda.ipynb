# 2. Phụ lục 

## 2.1 Quá trình phân tích dữ liệu theo các bước.

- Bước 1: Prepare: Thu thập, làm sạch và xử lý trước dữ liệu thô để phù hợp cho việc phân tích. 

- Bước 2: Xây dựng mô hình dữ liệu thể hiện cấu trúc, mối quan hệ và các ràng buộc của dữ liệu. 

- Bước 3: Phân tích: Cốt lõi của phân tích dữ liệu giúp đào sâu để khám phá những hiểu biết sâu sắc và trả lời câu hỏi.

- Bước 4: Trực quan: Trực quan hóa dữ liệu là một công cụ mạnh mẽ cho khám phá dữ liệu, xác định các mẫu và truyền đạt những phát hiện một cách hiệu quả. 

- Bước 5: Quản lý (manage): đảm bảo tính toàn vẹn, nhất quán, và bảo mật của dữ liệu đang được sử dụng. 

## 2.2 4 câu hỏi trong phân tích dữ liệu.

- Phân tích mô tả (Descriptive Analytics): Trả lời các câu hỏi về những gì đã xảy ra trong một khoảng thời gian nhất định bằng cách tóm tắt dữ liệu trong quá khứ và trình bày các kết quả. 

- Phân tích chuẩn đoán (Diagnostic Analytics): Những hiểu biết từ phân tích mô tả để đào sâu hơn để tìm ra nguyên nhân của kết quả. 

- Phân tích dự đoán (Predictive Analytics): Dữ liệu lịch sử và xu hướng được sử dụng để dự đoán. Lưu ý là mục đích **không phải để nói những gì sẽ xảy ra, đó là dự báo những gì có thể xảy ra**. 

- Phân tích đề xuất (Prescriptive Analytics): Phân tích các quyết định và sự kiện trong quá khứ, khả năng của các kết quả khác nhau.

# Phân tích về performance của những người bán hàng. 

Để phân tích được performance của những người bán hàng, có nhiều yếu tố mà ta cần phải quan tâm đến, có thể kể đến trong đó như: 

1. Tổng doanh thu theo người bán.

2. Tổng lợi nhuận theo người bán.

3. Tổng đơn đặt hàng theo người bán.

4. Giá trị trung bình của đơn theo người bán.

```python
data = reg_sale_all
```

```python
# Group data by Sales Agent
agent_performance = data.groupby('Sales Agent').agg({
    'Sale Price': 'sum',
    'Total Profit (GMROI)': 'sum',
    'Order ID': 'count'
}).reset_index()

# Rename columns for better clarity
agent_performance.columns = ['Sales Agent', 'Total Sales', 'Total Profit', 'Number of Orders']

import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(14, 8))

# Total Sales by Agent
plt.subplot(2, 1, 1)
plt.bar(agent_performance['Sales Agent'], agent_performance['Total Sales'], color='skyblue')
plt.title('Total Sales by Sales Agent')
plt.xlabel('Sales Agent')
plt.ylabel('Total Sales')

# Total Profit by Agent
plt.subplot(2, 1, 2)
plt.bar(agent_performance['Sales Agent'], agent_performance['Total Profit'], color='lightgreen')
plt.title('Total Profit by Sales Agent')
plt.xlabel('Sales Agent')
plt.ylabel('Total Profit')

plt.tight_layout()
plt.show()
```

Kế đến xem xét giá trung bình theo đơn của từng người bán

```python
# Calculate Average Order Value by Agent
agent_performance['Average Order Value'] = agent_performance['Total Sales'] / agent_performance['Number of Orders']

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(agent_performance['Sales Agent'], agent_performance['Average Order Value'], color='coral')
plt.title('Average Order Value by Sales Agent')
plt.xlabel('Sales Agent')
plt.ylabel('Average Order Value')
plt.show()
```

Ngoài ra chúng ta cũng có thể phân tích thêm về trend bán hàng theo thời gian của từng agent khác nhau. 

```python
# Extract year and month from the Order Date for aggregation
data['YearMonth'] = data['Order Date'].dt.to_period('M')

# Group data by Sales Agent and YearMonth to get monthly sales for each agent
monthly_sales_by_agent = data.groupby(['YearMonth', 'Sales Agent'])['Sale Price'].sum().unstack().fillna(0)

# Plotting the data
plt.figure(figsize=(14, 8))
for agent in monthly_sales_by_agent.columns:
    plt.plot(monthly_sales_by_agent.index.astype(str), monthly_sales_by_agent[agent], marker='o', label=agent)

plt.title('Monthly Sales Trends by Sales Agent')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(title='Sales Agent')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

Nhìn chung, biểu đồ chỉ ra rằng trong khi một số nhân viên như Priscilla và Kelly luôn có hiệu suất tốt, những người khác lại có hiệu suất biến đổi nhiều hơn, nhấn mạnh sự cần thiết của các chiến lược mục tiêu để tối ưu hóa doanh số trong suốt cả năm.


```python
from scipy import stats

# Perform t-tests between each pair of sales agents
agent_combinations = [(a, b) for idx, a in enumerate(monthly_sales_by_agent.columns) for b in monthly_sales_by_agent.columns[idx + 1:]]
t_test_results = {}

for agent1, agent2 in agent_combinations:
    t_stat, p_value = stats.ttest_ind(monthly_sales_by_agent[agent1], monthly_sales_by_agent[agent2], equal_var=False)
    t_test_results[(agent1, agent2)] = (t_stat, p_value)

# Display t-test results
t_test_results
```

Giả định đặt ra là có một vài mặt hàng mà nhân viên kinh doanh sẽ dựa vào đó mà bán hàng được tốt hơn. 

Do đó mà cần thực hiện phân tích thử tác động của danh mục sản phẩm lên khả năng bán được hàng của doanh nghiệp. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Group data by Product Category and Sales Agent
category_performance = reg_sale_all.groupby(['Product Category', 'Sales Agent']).agg({
    'Sale Price': 'sum',
    'Total Profit (GMROI)': 'sum'
}).unstack().fillna(0)

# Extract the product categories and sales agents
product_categories = category_performance.index
sales_agents = category_performance.columns.levels[1]

# Number of product categories
n_categories = len(product_categories)
# Number of agents
n_agents = len(sales_agents)

# Define the bar width and positions
bar_width = 0.8 / n_agents
bar_positions = np.arange(n_categories)

# Plotting the data
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12), sharex=True)

# Plot total sales by product category for each agent
for i, agent in enumerate(sales_agents):
    ax1.bar(bar_positions + i * bar_width, category_performance[('Sale Price', agent)], width=bar_width, label=agent)

ax1.set_title('Total Sales by Product Category and Sales Agent')
ax1.set_ylabel('Total Sales')
ax1.set_xticks(bar_positions + bar_width * (n_agents - 1) / 2)
ax1.set_xticklabels(product_categories, rotation=45)
ax1.legend(title='Sales Agent')

# Plot total profit by product category for each agent
for i, agent in enumerate(sales_agents):
    ax2.bar(bar_positions + i * bar_width, category_performance[('Total Profit (GMROI)', agent)], width=bar_width, label=agent)

ax2.set_title('Total Profit by Product Category and Sales Agent')
ax2.set_ylabel('Total Profit')
ax2.set_xticks(bar_positions + bar_width * (n_agents - 1) / 2)
ax2.set_xticklabels(product_categories, rotation=45)
ax2.legend(title='Sales Agent')

plt.xlabel('Product Category')
plt.tight_layout()
plt.show()
```

   - Tập trung vào cải thiện tỷ suất lợi nhuận cho các đại lý như Kelly bằng cách xem xét lại chi phí và chiến lược giá.

2. **Nâng cao đào tạo và hỗ trợ**:
   - Cung cấp đào tạo và hỗ trợ bổ sung cho các đại lý như Terrence để cải thiện hiệu suất tổng thể của họ.

3. **Tận dụng điểm mạnh**:
   - Khuyến khích các đại lý tập trung vào điểm mạnh của họ. Ví dụ, Daniel nên tiếp tục tập trung vào điện tử, trong khi Priscilla có thể tận dụng điểm mạnh của mình trên nhiều danh mục.

4. **Tăng giá trị đơn hàng trung bình**:
   - Thực hiện các chiến lược để tăng giá trị đơn hàng trung bình cho các đại lý như Brian.

5. **Tiếp thị và khuyến mãi mục tiêu**:
   - Phát triển các chiến dịch tiếp thị mục tiêu dựa trên điểm mạnh và điểm yếu của từng đại lý và hiệu suất của các danh mục sản phẩm khác nhau.

6. **Theo dõi xu hướng và điều chỉnh chiến lược**:
   - Liên tục theo dõi xu hướng doanh số và điều chỉnh chiến lược để tối ưu hóa hiệu suất trong suốt cả năm.
